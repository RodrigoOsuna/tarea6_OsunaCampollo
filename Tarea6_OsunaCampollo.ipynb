Seguir un procedimiento similar al mostrado cuando se generó una distribución binomial, pero en esta caso genere un código que genere variables aletorias Poisson cuya función de distribución de probabilidad esta dada por:
>$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$
    
> Demuestre matemáticamente que 
> $$P(k+1)=\frac{\lambda}{k+1}P(k)$$
> y a partir de esta relación genere variables aletorias que distribuyen poisson usando el método de la transformada inversa.

### Demostración:


$$P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!}$$

$$P(k + 1)=\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}$$

$$P(k + 1)=\frac{e^{-\lambda}(\lambda)^{k}\lambda}{(1+k)k!}$$

Sabiendo el valor de P(k,$ \lambda $)
$$P(k + 1)=\frac{\lambda}{(1+k)} P(k,\lambda) $$


```python
#Problema 1 forma con ecuacion
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sps
from math import factorial as fac
from scipy import optimize
import math

def poisson(n:"numero de iteraciones",y:"numero de ocurrencias por intervalo de tiempo"):
    k=np.arange(n) #variable que contiene todos los numeros de exitos posibles 
    t=np.exp(-y)*(y**k) #Primera parte de la ecuacion, por alguna razon no me salia si no lo separaba de esta manera
    def llenar(i:"iterador usado en la compresion de lista como argumento"):
        nonlocal t #Variable que vamos a modificar
        t[i]=t[i]/(sps.gamma(k[i]+1))#modificacion necesaria para terminar la formula
    [llenar(i) for i in range(len(k))]#mandamos llamar la funcion para que realize la modificacion en nuestro vector
    return t#regresamos el vector ya termiando
        
n=10
y=5 #numero de incidencias que hay por intervalo de tiempo
v=poisson(n,y)
t=np.arange(n)
plt.plot(t,v)#arreglo con probabilidades para cada k posible dentro del rango de incidencias
plt.show
```

```python
#Problema 3 Forma iterativa
def poisson_it(n,y:"numero de ocurrencias por intervalo de tiempo"):
    k=np.zeros(n)#variable vacia que contiene todos los numeros de exitos posibles
    k[0]=np.exp(-y) #Formula general usada para encontrar el primer valor de la lista
    def llenar(i:"iterador usado en la compresion de lista como argumento"):
        nonlocal k
        k[i+1] = (y/(i+1))*k[i]
    [llenar(i) for i in range(len(k)-1)]#mandamos llamar la funcion para que realize la modificacion en nuestro vector
    return k
y=50
n=100
v=poisson_it(n,y)
t=np.arange(n)
plt.plot(t,v)
plt.show()
```
